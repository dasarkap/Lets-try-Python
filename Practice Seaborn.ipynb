```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
#Jointplot
sns.jointplot(x=df["total_bill"],
            y=df["tip"],
            data=df)
```

```python

sns.jointplot(x=df["total_bill"],
            y=df["tip"],
            data=df,kind='hex')
```

```python
sns.jointplot(x=df["total_bill"],
            y=df["tip"],
            data=df,kind='kde')
```

```python
#Pairplot
sns.pairplot(df,hue='sex')
```

```python
#Bar Plot
sns.barplot(x=df["sex"],
           y=df['total_bill'])
```

```python
sns.barplot(x=df["sex"],
           y=df['tip'],data=df)
```

```python
#Count Plot
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(x='tip',data=df)
```

```python
sns.boxenplot(x=df["sex"],y=df["total_bill"])
```

```python
#strip plot
sns.stripplot(x=df["day"],y=df["tip"],data=df,hue='sex')
```

```python
#Heat Map
sns.heatmap(df.corr(),annot=True)
```

```python
sns.heatmap(df.corr(),annot=True,cmap='coolwarm',linecolor='white',linewidth=3)
```

```python
#Pair grid Plot
my_var=sns.PairGrid(df)
my_var.map_diag(sns.distplot)
my_var.map_upper(sns.kdeplot)
my_var.map_lower(sns.boxenplot)
```

```python
#Facet Grid
g=sns.FacetGrid(data=df,row='smoker')
g.map(sns.distplot,'total_bill')
```
