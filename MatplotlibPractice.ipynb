```python
import matplotlib.pyplot as mp
import seaborn as sns
```

```python
mp.plot([2,4,6,4])
mp.ylabel("Numbers")
mp.xlabel("Indices")
mp.title("Myplot")
mp.show()
```

```python
import numpy as np
t=np.arange(0.,5.,0.25)
mp.plot(t,t**2,'b--',label='^2')
mp.plot(t,t**2.2,'rs',label='^2.2')
mp.plot(t,t**2.5,'g^',label='^2.5')
mp.grid()
mp.legend()
mp.show()
```

```python
x1=[1,2,3,4]
x2=[4,6,8,3]
y1=[3,6,9,12]
y2=[2,3,4,1.5]
lines=mp.plot(x1,x2,y1,y2)
lines=mp.plot(x1,y1,x2,y2)
mp.setp(lines[0],color='r',linewidth=2.0)
mp.setp(lines[1],color='g',linewidth=2.0)
mp.grid()
```

```python
from matplotlib import style
x=[2,3,2]
y=[4,2,3]
fig=mp.figure()
ax1=fig.add_subplot(211)
style.use('ggplot')
ax1.plot(x,y)
mp.plot(x,y,label='my First Graph')
mp.xlabel('X axis')
mp.ylabel('Y axis')
mp.title('My Graph\n Arka')
mp.show()

ax2=fig.add_subplot(212)
style.use('dark_background')
ax2.plot(x,y)
mp.plot(x,y,label='my First Graph')
mp.xlabel('X1 axis')
mp.ylabel('Y1 axis')
mp.title('My Graph\n Arka1')
mp.show()
```
